```python
import pandas as pd 
import os
from datetime import datetime, timedelta

ticker = 'NQ-'
path = r'C:\Users\ianaz\Documents\Upwork projects\Miracle_fund\3-EMA-Strategy\NQ_min_raw'


df_list = []
for file in os.listdir(path):
    df_list.append(os.path.join(path, file))
    
print(df_list)
df_list = [s for s in df_list if (ticker in s)]
print(df_list)

QQQ_whole = pd.DataFrame()
for i in df_list:
    print(i)
    df_temp = pd.read_csv(i)
    df_temp.set_index('DateTime', inplace=True)
    df_temp.index = pd.to_datetime(df_temp.index)
    shift = df_temp.index[0].hour - 9
    print(shift)
    df_temp.index = df_temp.index - timedelta(hours=shift)
    
    QQQ_whole = QQQ_whole.append(df_temp)
    
    display(df_temp.head())

QQQ_whole.drop(columns='Unnamed: 0', inplace=True)

# QQQ_whole.set_index('DateTime', inplace=True)
# QQQ_whole.index = pd.to_datetime(QQQ_whole.index)
QQQ_whole = QQQ_whole.sort_index()
QQQ_whole.drop_duplicates(inplace=True)
QQQ_whole = QQQ_whole[~QQQ_whole.index.duplicated(keep='first')]

QQQ_whole.to_csv(r'C:\Users\ianaz\Documents\Upwork projects\Miracle_fund\3-EMA-Strategy\NQ_min_raw\NQ_whole.csv')
```
